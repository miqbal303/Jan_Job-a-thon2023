# Converting categorical into numerical

```python
train['area'] = train['area'].map({'Urban': 0 , 'Rural':1})
train['qualification'] = train['qualification'].map({'High School':0,'Bachelor':1,'Others':2})
train['income'] = train['income'].map({'5L-10L':0,'2L-5L':1,'More than 10L':2,'<=2L':3})
train['num_policies'] = train['num_policies'].map({'More than 1':0,'1':1})
train['policy'] = train['policy'].map({'A':0,'B':1,'C':2})
train['type_of_policy'] = train['type_of_policy'].map({'Platinum':0,'Silver':1,'Gold':2})
```

```python
test['area'] = test['area'].map({'Urban': 0 , 'Rural':1})
test['qualification'] = test['qualification'].map({'High School':0,'Bachelor':1,'Others':2})
test['income'] = test['income'].map({'5L-10L':0,'2L-5L':1,'More than 10L':2,'<=2L':3})
test['num_policies'] = test['num_policies'].map({'More than 1':0,'1':1})
test['policy'] = test['policy'].map({'A':0,'B':1,'C':2})
test['type_of_policy'] = test['type_of_policy'].map({'Platinum':0,'Silver':1,'Gold':2})
```

# Trains Test Split

```python
X = train.iloc[:,:-1]
y = train['cltv']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=75)
```

# ML algorithm LGBM

```python
import lightgbm 
from lightgbm import LGBMRegressor
```

```python
# Define the model
lgbm = LGBMRegressor(boosting_type=  'gbdt',
    num_leaves = 80,
    max_depth = 3,
    learning_rate = 0.1,
    n_estimators = 100,
    subsample_for_bin = 30000,
    min_split_gain= 0.1,
    min_child_weight = 1e-1,
    min_child_samples =10,
    subsample = 1,
    subsample_freq = 2,
    colsample_bytree = 0.6,
    reg_alpha = 0.05,
    reg_lambda = 0.05,
    random_state = 42,
    n_jobs= -1,
)
```

```python
# Fit data
lgbm.fit(X_train, y_train)
```

```python
y_pred = lgbm.predict(X_test)
```

```python
#R2 score
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
